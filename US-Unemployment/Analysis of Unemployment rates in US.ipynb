We see that all the counties have 363 records each, Oregon State has 531 records and National has 868 records.  This means that not all files have data for all the years. By displaying the first few records of each file will show us the starting year of each file. National starts at 1948, Oregon starts at 1976 and Counties (All) start at 1990. So we'll only consider the period after 1990 from the files for better insights on the relationship between the entities.

```python
df_National['DATE'] = pd.to_datetime(df_National['DATE'], format='%Y-%m-%d')
df_Oregon['DATE'] = pd.to_datetime(df_Oregon['DATE'], format='%Y-%m-%d')
df_Jackson['DATE'] = pd.to_datetime(df_Jackson['DATE'], format='%Y-%m-%d')
df_Coos['DATE'] = pd.to_datetime(df_Coos['DATE'], format='%Y-%m-%d')
df_Curry['DATE'] = pd.to_datetime(df_Curry['DATE'], format='%Y-%m-%d')
df_Douglas['DATE'] = pd.to_datetime(df_Douglas['DATE'], format='%Y-%m-%d')
df_Klamath['DATE'] = pd.to_datetime(df_Klamath['DATE'], format='%Y-%m-%d')
df_Josephine['DATE'] = pd.to_datetime(df_Josephine['DATE'], format='%Y-%m-%d')
```

```python
df = df_National
df.dtypes
```

```python
df_National90 = df_National[(df_National['DATE'].dt.year >= 1990)]
df_Oregon90 = df_Oregon[(df_Oregon['DATE'].dt.year >= 1990)]
```

# Plot line graph

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20,16)
plt.plot(df_National90['DATE'],df_National90['UNRATE'],color="black", label="National")
plt.plot(df_Oregon90['DATE'],df_Oregon90['ORUR'],color="magenta", label="Oregon")#linewidth=2.0)
plt.plot(df_Jackson['DATE'],df_Jackson['ORJACK0URN'],color="green", label="Jackson")#linewidth=2.0)
plt.plot(df_Coos['DATE'],df_Coos['ORCOOS1URN'],color="cyan", label="Coos")#linewidth=2.0)
plt.plot(df_Curry['DATE'],df_Curry['ORCURR5URN'],color="orange", label="Curry")#linewidth=2.0)
plt.plot(df_Douglas['DATE'],df_Douglas['ORDOUG5URN'],color="red", label="Douglas")#linewidth=2.0)
plt.plot(df_Klamath['DATE'],df_Klamath['ORKLAM5URN'],color="yellow", label="Klamath")#linewidth=2.0)
plt.plot(df_Josephine['DATE'],df_Josephine['ORJOSE5URN'],color="blue", label="Josephine")#linewidth=2.0)
plt.legend(loc="upper left")

plt.show()
```
